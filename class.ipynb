## Class and Object

```python
class Parrot:

    # class attribute
    species = "bird"

    # instance attribute
    def __init__(self, name, age):
        self.name = name
        self.age = age

# instantiate the Parrot class
blu = Parrot("Blu", 10)
woo = Parrot("Woo", 15)

# access the class attributes
print("Blu is a {}".format(blu.__class__.species))
print("Woo is also a {}".format(woo.__class__.species))

# access the instance attributes
print("{} is {} years old".format( blu.name, blu.age))
print("{} is {} years old".format( woo.name, woo.age))
```

## Methods

```python
class Parrot:
    
    # instance attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # instance method
    def sing(self, song):
        return "{} sings {}".format(self.name, song)

    def dance(self):
        return "{} is now dancing".format(self.name)

# instantiate the object
blu = Parrot("Blu", 10)

# call our instance methods
print(blu.sing("'Happy'"))
print(blu.dance())
```

```python
#self가 들어가면 instance method
```

```python
##self를 다른 거로 바꿔도 상관없음
#즉, 메소드의 첫번째 인자는 자기자신을 가리키는 것! self는 관례상 이름.. 

class Parrot: 
    
    # instance attributes
    def __init__(god, name, age):
        god.name = name
        god.age = age
    
    # instance method
    def sing(god, song):
        return "{} sings {}".format(god.name, song)

    def dance(god):
        return "{} is now dancing".format(god.name)
    
    @classmethod
    def dance2(cls):
        return "{} is now dancing".format(cls)

# instantiate the object
blu = Parrot("Blu", 10)

# call our instance methods
print(blu.sing("'Happy'"))
print(blu.dance())
```

## Inheritance

```python
# parent class
class Bird:
    
    def __init__(self):
        print("Bird is ready")

    def whoisThis(self):
        print("Bird")

    def swim(self):
        print("Swim faster")

# child class
class Penguin(Bird):
    
    def swim(self):
        print("I Can Swim")

    
    def __init__(self):
        # call super() function
        super().__init__() #원래 super는 부모가 가진 것을 실행하라는 뜻인데..
#         super(Bird).__init__() 
#         super(Penguin).__init__() 
#         super(Bird, self).__init__() #python의 super는 type을 넣을 수 있음 >> 다중상속!!!
#         super(Penguin, self).__init__()
        print("Penguin is ready")

#     def whoisThis(self):
#         print("Penguin")

    def run(self):
        print("Run faster")

peggy = Penguin()
peggy.whoisThis()
peggy.swim()
peggy.run()
```

```python
class Polygon:
    def __init__(self, no_of_sides):
        self.n = no_of_sides #self.n, self.sides는 인스턴스 어트리뷰트
        self.sides = [0 for i in range(no_of_sides)] #컴프리헨션 안에는 인수 아무거나 다 쓸 수 있다

    def inputSides(self):
        self.sides = [float(input("Enter side "+str(i+1)+" : ")) for i in range(self.n)]

    def dispSides(self):
        for i in range(self.n):
            print("Side",i+1,"is",self.sides[i])
```

```python
class Triangle(Polygon):
    def __init__(self):
        Polygon.__init__(self,3) # super().__init__(3)

    def findArea(self):
        a, b, c = self.sides
        # calculate the semi-perimeter
        s = (a + b + c) / 2
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        print('The area of the triangle is %0.2f' %area)
```

```python
#나만의 INT 덧셈을 오버로딩과 오버라이딩 이용해서 만들기! 사실 앞에 플러스, 뒤에 플러스 나올 때, += 까지 다 만들어야...
class MyINT(int):
    def __init__(self, num):  #self는 객체이지 값이 아닌 점을 주의!
        self.num = num
        
    def __add__(self, other): #self는 객체이지 값이 아닌 점을 주의!
        print('와하우')
        return (self.num%10)+(other%10)
    def __radd__(self, other): #radd는? 오른쪽편에 플러스가 있으면 radd!
        print('알하우')
        return (self.num%10)+(other%10)
    
    def __sub__(self, other):
        print('와하우')
        return (self.num%10)-(other%100)
    def __rsub__(self, other):
        print('알하우')
        return (self.num%10)-(other%100)
    
    def __str__(self):
        print('개나이스')
        return str(self.num)
    def __repr__(self):
        return('왜 안나와') 
```

```python
class Point:
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y
    
    def __str__(self):
        return "({0},{1})".format(self.x,self.y)
    
    def __lt__(self,other):
        self_mag = (self.x ** 2) + (self.y ** 2)
        other_mag = (other.x ** 2) + (other.y ** 2)
        return self_mag < other_mag
```
