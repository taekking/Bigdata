```python
def getParse(url, params=None, num_retries=2):
    headers = {"User-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_5) \
               AppleWebKit/537.36"
              }   
    req = requests.get(url, params, headers=headers)
    
    if 500 <= req.status_code < 600 and num_retries > 0:
        print("error: ", req.status_code, req.reason)
        return getParse(url, num_retries=(num_retries))
    
    result = BeautifulSoup(req.content, "lxml")
    
    return result
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('city_air.db')
```

```python
cur = conn.cursor()
```

```python
cur.execute("create table city_air (datatime, so2value, covalue, o3value, no2value, pm10value, pm10value24, pm25value, pm25value24, khaivalue, khagrade, so2grade, cograde, o3grade, no2grade, pm10grade, pm25grade, pm10grade1h, pm25grade1h)")
```

```python
curData = []

for link in result.select("item"):
    Tuple = tuple()
    a = link.find_next()
    
    for i in range(len(link.findAll())):
        if a.name != 'mangname':
            Tuple = Tuple + (a.get_text(),)
        a = a.find_next()
    
    curData.append(Tuple)
```

```python
conn.executemany("insert into city_air (datatime, so2value, covalue, o3value, no2value, pm10value, pm10value24, pm25value, pm25value24, khaivalue, khagrade, so2grade, cograde, o3grade, no2grade, pm10grade, pm25grade, pm10grade1h, pm25grade1h) \
values (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)", curData);
```
